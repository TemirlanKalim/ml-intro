**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
clear_data=match_data.loc[(match_data["league_id"]==Eng_Id) & (match_data["season"]=="2010/2011")].loc[:,['date','match_api_id','home_team_api_id','away_team_api_id']].sort_values(by=["date","match_api_id"])
```

```python
clear_data
```

```python
highest_gap=pd.DataFrame(columns=["date","team"])
i=0
for _,x in clear_data.iterrows():
  highest_gap.loc[i]=[x.date,x.away_team_api_id]
  i=i+1
  highest_gap.loc[i]=[x.date,x.home_team_api_id]
  i=i+1
highest_gap
```

```python
highest_gap2=highest_gap.sort_values(['team','date'])\
                                     .reset_index(drop=True)
```
